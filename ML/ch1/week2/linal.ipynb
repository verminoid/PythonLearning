```python
data = []
with open('sentences.txt', 'r') as f:
    d = f.readlines()
    for s in d:
        data.append(s.lower())
```

```python
import re
los = []
for st in data:
    st_spl = re.split('[^a-z]', st)
    los.append(list(filter(str.strip, st_spl)))
los
```

```python
all_w = {}
for s in los:
    for w in s:
        all_w.update({w:''})
l = list(enumerate(all_w))
all = {}
for n, w in l:
    all.update({w:n})
all
```

```python
import numpy as np

lin = np.zeros((len(los),len(all)))
for i in range(len(los)):
    for w in los[i]:
        t = np.zeros((len(los),len(all)))
        t[i,all[w]] = 1
        lin = lin + t



```

```python
from scipy.spatial import distance

d = [1,]
for i in range(1,len(lin)):
    d.append(distance.cosine(lin[0],lin[i]))
a = d.copy()
a.sort()
print(d.index(a[0]),d.index(a[1]))
```
