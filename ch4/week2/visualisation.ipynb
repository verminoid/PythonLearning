```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('CompleteDataset.csv')
```

```python
df.columns
```

```python
len(df.groupby(['Club']).groups)
```

```python
lgr = []
for name, group in df.groupby(['Age']):
    lgr.append(name)
print(lgr)
```

```python
N= len(df.groupby(['Age']).groups)
cmap = cmap =plt.cm.get_cmap('gist_rainbow', N)
fig, ax = plt.subplots(figsize=(15,10))
i=0
for name, group in df.groupby(['Age']):
    group.plot.hist('Strength', color=cmap(i), label=name, ax=ax)
    i += 1
```
